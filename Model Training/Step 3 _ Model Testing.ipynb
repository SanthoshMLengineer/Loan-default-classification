```python
import numpy as  np
import pandas as pd

dict_test_data = {
     	 	 	
    "Gender" : "Male",
    "Married" : "Yes",
    "Dependents" : "0",
    "Education" : "Graduate",
    "Self_Employed" : "Yes",
    "LoanAmount" : 66,
    "Total_income" : 3000,
    "Loan_Amount_Term" : 360.0,
    "Credit_History" : 1.0,
    "Property_Area" :"Urban",
}

data_input = pd.DataFrame(dict_test_data, index = [0])

data_input
```

```python
data_input['LoanAmount_log'] = np.log(data_input['LoanAmount'])
```

```python
data_input['Dependents'] = data_input['Dependents'].map({"0":0,
                                                         "1":1,"2":2,"3+":3})

data_input['Education'] = data_input['Education'].map({"Graduate":1,
                                                       "Not Graduate":0})

data_input['Gender'] = data_input['Gender'].map({"Male":1,"Female":0})

data_input['Married'] = data_input['Married'].map({"Yes":1,"No":0})

data_input['Self_Employed'] = data_input['Self_Employed'].map({"Yes":1,"No":0})

data_input['Property_Area'] = data_input['Property_Area'].map({"Rural":0,
                                                               "Semiurban":1,
                                                               'Urban':2})

data_input
```

```python
data_input['LoanAmount_log'] = np.log(data_input['LoanAmount'])
data_input.drop(['LoanAmount'], axis = 1, inplace = True)
data_input
```
